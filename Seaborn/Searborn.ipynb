## ðŸ”— Relplot - Figure Level Function

> **A unified interface for creating relational plots (scatter & line)**

### ðŸŽ¯ What is Relplot?

**Relplot** is a figure-level function that creates an entire figure with relationship plots. It automatically handles figure creation and subplot management.

---

### ðŸ“Š Figure Level vs Axis Level

| Aspect | **Figure Level** (`relplot`) | **Axis Level** (`scatterplot`) |
|--------|------------------------------|--------------------------------|
| **Shape** | Square-shaped by default | Rectangular, follows axes shape |
| **Creates** | Entire figure + axes | Only draws on existing axes |
| **Subplots** | Automatic grid creation (`col`, `row`) | Manual setup required |
| **Flexibility** | Best for quick, multi-panel plots | Best for custom layouts |
| **Use Case** | Exploratory analysis, faceting | Precise control, dashboards |

---

### ðŸ’» Code Breakdown

```python
sns.relplot(
    data=tips,           # Dataset to plot
    x='total_bill',      # X-axis variable
    y='tip',             # Y-axis variable
    kind='scatter',      # Plot type: 'scatter' or 'line'
    hue='sex',           # Color points by category
    style='time',        # Different markers per category
    size='size'          # Point size based on value
)
```

## ðŸ“ˆ KDE Plot - Kernel Density Estimation

> **A smooth, continuous curve showing data distribution probability**

### ðŸŽ¯ What is KDE Plot?

**KDE (Kernel Density Estimation)** creates a smooth probability density curve by placing a Gaussian (bell-shaped) kernel over each data point and summing them. It's like a smoothed histogram.

**Simple Analogy:** Imagine placing a small bell curve on each data point, then blending them all together to create one smooth wave.

---

### ðŸ“Š Histogram vs KDE Plot

| Aspect | **Histogram** | **KDE Plot** |
|--------|---------------|--------------|
| **Shape** | Bars (discrete bins) | Smooth curve (continuous) |
| **Appearance** | Blocky, jagged | Flowing, elegant |
| **Bin Dependency** | Sensitive to bin size | No bins needed |
| **Best For** | Exact counts | Overall distribution shape |

---

### ðŸ’» Code Breakdown

```python
sns.kdeplot(
    data=tips,          # Dataset
    x='total_bill'      # Variable to plot
)
```

#### Parameter Details

| Parameter | Required | Description | Example Values |
|-----------|----------|-------------|----------------|
| `data` | âœ… Yes | DataFrame containing the data | `tips`, `df` |
| `x` | âœ… Yes | Column for distribution | `'total_bill'`, `'tip'` |
| `y` | âŒ No | Second variable (creates 2D density) | `'tip'` |
| `hue` | âŒ No | Color by category | `'sex'`, `'day'` |
| `fill` | âŒ No | Fill area under curve | `True`, `False` |
| `bw_adjust` | âŒ No | Smoothness control (0.5 = smoother, 2 = rougher) | `0.5`, `1`, `2` |
| `cut` | âŒ No | Extend curve beyond data range | `0`, `2` (default) |
| `cumulative` | âŒ No | Show cumulative distribution | `True`, `False` |

## ðŸ“ Rug Plot - Marginal Distribution Markers

> **Small ticks along axes showing exact location of each data point**

### ðŸŽ¯ What is Rug Plot?

A **rug plot** adds small vertical (or horizontal) tick marks along the axis edges, where each tick represents one individual data point. It's like a minimalist scatter plot on the margin.

**Purpose:** Complement other plots (KDE, histogram) by showing the actual raw data locations without cluttering the main visualization.

---

### ðŸ“Š Why Use Rug Plot?

| Benefit | Description |
|---------|-------------|
| **Shows Raw Data** | See exact positions of individual observations |
| **Unobtrusive** | Doesn't interfere with main plot |
| **Reveals Gaps** | Identify where no data exists |
| **Complements KDE** | Verify if smooth curve matches actual data density |

---

### ðŸ’» Code Breakdown

```python
# Main distribution plot
sns.kdeplot(
    data=tips,
    x='total_bill'
)

# Add individual data markers
sns.rugplot(
    data=tips,
    x='total_bill'
)
```

| Parameter | Required | Description | Example Values |
|-----------|----------|-------------|----------------|

## ðŸ“Š Bivariate Histogram - Two-Variable Distribution

> **Shows the relationship between two variables using colored rectangular bins**

### ðŸŽ¯ What is Bivariate Histogram?

A **bivariate histogram** divides the 2D plot area into rectangular grids (bins) and uses color intensity to show how many data points fall into each rectangle. It's like a regular histogram but in two dimensions.

**Simple Analogy:** Imagine dividing a room into floor tiles, then coloring each tile based on how many people are standing on it.

---

### ðŸ“Š Univariate vs Bivariate Histogram

| Aspect | **Univariate** | **Bivariate** |
|--------|----------------|---------------|
| **Variables** | 1 (e.g., `total_bill`) | 2 (e.g., `total_bill` + `tip`) |
| **Shape** | Bars along one axis | Color-coded rectangles |
| **Shows** | Frequency distribution | Joint distribution & relationship |
| **Axes** | Values + Count | Variable 1 + Variable 2 |

---

### ðŸ’» Code Breakdown

```python
sns.displot(
    data=tips,           # Dataset
    x='total_bill',      # First variable (x-axis)
    y='tip',             # Second variable (y-axis)
    kind='hist'          # Histogram type
)
```

#### Parameter Details

| Parameter | Required | Description | Example Values |
|-----------|----------|-------------|----------------|
| `data` | âœ… Yes | DataFrame containing the data | `tips`, `df` |
| `x` | âœ… Yes | Variable for x-axis | `'total_bill'` |
| `y` | âœ… Yes | Variable for y-axis (for bivariate) | `'tip'` |
| `kind` | âŒ No | Distribution plot type | `'hist'`, `'kde'`, `'ecdf'` |
| `bins` | âŒ No | Number of bins for each axis | `10`, `20`, `(10, 15)` |
| `cbar` | âŒ No | Show colorbar legend | `True`, `False` |
| `hue` | âŒ No | Color by category | `'sex'`, `'day'` |

## ðŸŽ¯ Matrix Plots - Visualizing Tabular Data

### ðŸ”¥ Heatmap

**A color-coded grid where each cell's color represents its numerical value**

#### What It Does:
- Displays matrix/table data using color intensity
- Darker/brighter colors = higher/lower values

#### Best For:
- Correlation matrices between variables
- Confusion matrices in ML models
- Time-series data (days Ã— hours)
- Comparison across two categorical dimensions

---

### ðŸŒ³ Clustermap

**An advanced heatmap with hierarchical clustering that automatically reorders rows/columns**

#### What It Does:
- Creates a heatmap PLUS dendrograms (tree diagrams)
- Automatically groups similar rows and columns together
- Reorders data to reveal hidden patterns

#### Key Features:
- **Dendrograms** - Tree diagrams showing similarity relationships
- **Auto-Grouping** - Similar items placed together
- **Pattern Discovery** - Reveals clusters you might miss
- **Hierarchical Structure** - Shows relationships at multiple levels

#### Best For:
- Gene expression data analysis
- Customer segmentation patterns
- Feature correlation with grouping
- Finding natural clusters in data

## ðŸ”¥ Heatmap - Color-Encoded Matrix Visualization

> **Transforms tabular data into a color-coded grid for instant pattern recognition**

---

### ðŸŽ¯ What Does It Do?

Converts a rectangular matrix (like pivot tables) into a visual grid where each cell's color represents its numerical value. Perfect for spotting trends, patterns, and outliers across two dimensions.

---

### ðŸ’» Code Breakdown

```python
# Step 1: Prepare data in matrix format

    index='country',      # Rows
    columns='year',       # Columns
    values='lifeExp'      # Cell values
)

# Step 2: Create heatmap
plt.figure(figsize=(15, 15))   # Set plot size
sns.heatmap(temp_df)            # Draw heatmap
```
