# üìä Basic Charts in Data Visualization

## Chart Types Overview

### üîπ Bar Chart
**Purpose:** Compare values across different categories

**When to use:**
- Comparing quantities between distinct groups
- Showing rankings or differences
- When category names are long

**Example:**
```
Sales by Product
Laptops:    ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà (120 units)
Phones:     ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà (80 units)
Tablets:    ‚ñà‚ñà‚ñà‚ñà‚ñà (50 units)
```

---

### üîπ Pie Chart
**Purpose:** Show parts of a whole (proportions)

**When to use:**
- Displaying percentage distribution
- When you have 3-6 categories max
- Total adds up to 100%

**Example:**
```
Budget Allocation
üü¶ Marketing: 40%
üü® Development: 35%
üü• Operations: 25%
```

‚ö†Ô∏è **Avoid when:** You have many categories or need precise comparisons

---

### üîπ Histogram
**Purpose:** Show distribution of continuous numerical data

**When to use:**
- Understanding data frequency
- Identifying patterns in measurements
- Finding the most common value ranges

**Example:**
```
Student Test Scores
0-20:   ‚ñÇ (5 students)
20-40:  ‚ñÑ (15 students)
40-60:  ‚ñà‚ñà‚ñà‚ñà (45 students)  ‚Üê Most common
60-80:  ‚ñÜ (25 students)
80-100: ‚ñÉ (10 students)
```

**Before:** Individual scores ‚Üí `45, 67, 89, 52, 78, 55, 61...`  
**After:** Grouped into ranges showing the pattern

---

## üìê What is Data Spread?

**Data spread** measures how scattered or concentrated your values are.

### Key Concepts:

**Range:** Distance between smallest and largest value
- Example: Test scores from 35 to 95 ‚Üí Range = 60

**Variance:** How far each value is from the average
- Low spread: Values clustered together (65, 67, 68, 70)
- High spread: Values scattered widely (20, 45, 70, 95)

**Why it matters:**
- Helps identify outliers (unusual values)
- Shows consistency in your data
- Reveals patterns and trends

---

## üéØ Quick Decision Guide

| Your Data | Best Chart |
|-----------|------------|
| Categories with quantities | **Bar Chart** |
| Percentages of a whole | **Pie Chart** |
| Continuous number ranges | **Histogram** |

**Pro Tip:** Histograms show "how many" in each range, Bar charts show "how much" for each category

# üìä Bar Chart - Complete Guide

## What is a Bar Chart?

A **bar chart** is used for quick comparison between categories. Each bar's length is directly proportional to its value.

**Simple Rule:** Big bar = Big value üìè

---

## üì¶ Sample Dataset

**Products vs Sales**

| Product | Sales (‚Çπ) |
|---------|-----------|
| A | 1000 |
| B | 1500 |
| C | 800 |
| D | 1200 |

---

## üîß Syntax Breakdown

```python
plt.bar(x, height, width, label)
```

### Parameters Explained

| Parameter | Type | Required | Description | Example |
|-----------|------|----------|-------------|---------|
| `x` | array/list | ‚úÖ **Yes** | Position of bars on x-axis (categories) | `['A', 'B', 'C', 'D']` |
| `height` | array/list | ‚úÖ **Yes** | Height of each bar (values/sales) | `[1000, 1500, 800, 1200]` |
| `width` | float | ‚ùå No | Thickness of bars (default: 0.8) | `0.5` (thinner), `1.0` (wider) |
| `label` | string | ‚ùå No | Name for legend | `'Sales Data'` |

---

## üíª Professional Implementation

```python
import matplotlib.pyplot as plt

# Data
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [1000, 1500, 800, 1200]

# Create bar chart
plt.bar(products, sales, width=0.6, label='Q1 Sales')

# Styling
plt.xlabel('Products')
plt.ylabel('Sales (‚Çπ)')
plt.title('Product Sales Comparison')
plt.legend()
plt.grid(axis='y', alpha=0.3)

# Display
plt.show()
```
---

## üé® Grid Styling

```python
plt.grid(axis='y', alpha=0.3)
```

### Parameters Explained

| Parameter | Type | Required | Description | Example |
|-----------|------|----------|-------------|---------|
| `axis` | string | ‚ùå No | Which axis to show gridlines | `'x'`, `'y'`, `'both'` (default) |
| `alpha` | float | ‚ùå No | Transparency level (0=invisible, 1=solid) | `0.3` (light), `0.7` (visible) |

**Why use grid?**
- Makes it easier to read exact values
- Helps compare bar heights quickly
- `alpha=0.3` keeps grid subtle and non-distracting

**Without Grid:**
```
Bars only - harder to estimate exact values
```

**With Grid (alpha=0.3):**
```
Bars with light horizontal lines - easy to read values at 1000, 1200, 1500
```

---

## üéØ Key Takeaways

‚úÖ Bar length is **directly proportional** to value  
‚úÖ Perfect for **comparing categories**  
‚úÖ Only `x` and `height` are mandatory  
‚úÖ Use `width` to adjust bar thickness (0.1 to 1.0)  
‚úÖ Use `label` when showing multiple datasets  
‚úÖ Add `grid` with low `alpha` for better readability

````python
import matplotlib.pyplot as plt
````

````python
# Data
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [1000, 1500, 800, 1200]

# Create bar chart
plt.bar(products, sales, width=0.4, label='Q1 Sales', color = 'orange')

# Styling
plt.xlabel('Products')
plt.ylabel('Sales (‚Çπ)')
plt.title('Product Sales Comparison')
plt.legend()
plt.grid(alpha = 0.3, axis='y')

# Display
plt.show()
````

# ü•ß Pie Chart - Complete Guide

## What is a Pie Chart?

A **pie chart** shows how different parts make up a whole. Each slice represents a proportion (percentage) of the total.

**Simple Rule:** Bigger slice = Bigger share üç∞

---

## üì¶ Sample Dataset

**Company Profit by Region**

| Region | Profit (‚Çπ) |
|--------|------------|
| North | 3000 |
| South | 2000 |
| East | 18000 |
| West | 1000 |

**Total Profit:** ‚Çπ24,000

---

## üîß Syntax Breakdown

```python
plt.pie(values, labels, colors, autopct, startangle, explode, shadow)
```

### Parameters Explained

| Parameter | Type | Required | Description | Example |
|-----------|------|----------|-------------|---------|
| `values` | array/list | ‚úÖ **Yes** | Data values (profit amounts) | `[3000, 2000, 18000, 1000]` |
| `labels` | array/list | ‚ùå No | Name for each slice | `['North', 'South', 'East', 'West']` |
| `colors` | array/list | ‚ùå No | Custom colors for slices | `['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A']` |
| `autopct` | string/function | ‚ùå No | Display percentages on slices | `'%1.1f%%'` (shows: 12.5%) |
| `startangle` | int | ‚ùå No | Rotation angle (default: 0) | `90` (starts from top) |
| `explode` | array/tuple | ‚ùå No | Separate slice from center | `(0, 0, 0.1, 0)` (pulls 3rd slice) |
| `shadow` | bool | ‚ùå No | Add shadow effect | `True` or `False` |

---

## üî¢ Understanding `autopct`

**autopct** automatically displays percentages on pie slices.

### Format String Breakdown

| Format | Meaning | Output Example |
|--------|---------|-----------------|
| `'%1.1f%%'` | 1 decimal place | `12.5%` |
| `'%1.2f%%'` | 2 decimal places | `12.50%` |
| `'%.0f%%'` | No decimals (whole number) | `13%` |
| `'%d%%'` | Integer only | `12%` |

**Explanation:**
- `%1.1f` = Show 1 digit before decimal, 1 digit after
- `%%` = Display the % symbol (first % escapes the second)

---

## üíª Professional Implementation

```python
import matplotlib.pyplot as plt

# Data
regions = ['North', 'South', 'East', 'West']
profits = [3000, 2000, 18000, 1000]
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A']

# Highlight the largest slice (East)
explode = (0, 0, 0.1, 0)

# Create pie chart
plt.pie(profits,
        labels=regions,
        colors=colors,
        autopct='%1.1f%%',
        startangle=90,
        explode=explode,
        shadow=True)

# Styling
plt.title('Regional Profit Distribution')
plt.axis('equal')  # Makes pie circular

# Display
plt.show()
```
---

## üéØ Common Use Cases

| autopct Format | Best For | Example Output |
|----------------|----------|----------------|
| `'%1.1f%%'` | **General use** (standard) | `12.5%` |
| `'%.0f%%'` | Simple/clean look | `13%` |
| `'%1.2f%%'` | Precise analysis | `12.50%` |

---

## ‚ö†Ô∏è Best Practices

‚úÖ Use for **3-6 categories** maximum  
‚úÖ All values must be **positive**  
‚úÖ Set `startangle=90` to start from top  
‚úÖ Use `explode` to highlight important slices  
‚úÖ Always use `autopct` to show percentages  
‚ùå Avoid for too many categories (use bar chart instead)

````python
# Data
regions = ['North', 'South', 'East', 'West']
profits = [3000, 2000, 18000, 1000]
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A']

# Highlight the largest slice (East)
explode = (0, 0, 0.1, 0)

# Create pie chart
plt.pie(profits,
        labels=regions,
        colors=colors,
        autopct='%1.1f%%',
        startangle=90,
        explode=explode,
        shadow=True)

# Styling
plt.title('Regional Profit Distribution')
plt.axis('equal')  # Makes pie circular

# Display
plt.show()
````

# üìä Histogram - Complete Guide

## What is a Histogram?

A **histogram** shows the distribution of continuous numerical data by grouping values into ranges (bins). It reveals patterns like which ranges are most common.

**Simple Rule:** Taller bar = More values in that range üìà

---

## üì¶ Sample Dataset

**100 Students Exam Scores**

| Score Range | Number of Students |
|-------------|-------------------|
| 0-20 | 5 |
| 20-40 | 12 |
| 40-60 | 28 |
| 60-80 | 35 |
| 80-100 | 20 |

**Raw Data:** `[45, 67, 89, 52, 78, 55, 61, 73, 82, 91, ...]` (100 scores)

---

## üîß Syntax Breakdown

```python
plt.hist(data, bins, color, edgecolor, alpha, range, density)
```

### Parameters Explained

| Parameter | Type | Required | Description | Example |
|-----------|------|----------|-------------|---------|
| `data` | array/list | ‚úÖ **Yes** | All individual values (student scores) | `[45, 67, 89, 52, 78, ...]` |
| `bins` | int/array | ‚ùå No | Number of ranges (default: 10) | `5` (5 ranges) or `[0,20,40,60,80,100]` |
| `color` | string | ‚ùå No | Bar fill color | `'skyblue'`, `'#45B7D1'` |
| `edgecolor` | string | ‚ùå No | Bar border color | `'black'`, `'white'` |
| `alpha` | float | ‚ùå No | Transparency (0=invisible, 1=solid) | `0.7` (semi-transparent) |
| `range` | tuple | ‚ùå No | Min and max values to include | `(0, 100)` |
| `density` | bool | ‚ùå No | Show as probability (default: False) | `True` (normalizes to 1.0) |

---

## üî¢ Understanding `bins`

**bins** determines how data is grouped into ranges.

### Bin Options

| Bins Value | Meaning | Result |
|------------|---------|--------|
| `5` | 5 equal ranges | `[0-20, 20-40, 40-60, 60-80, 80-100]` |
| `10` | 10 equal ranges | `[0-10, 10-20, 20-30, ..., 90-100]` |
| `[0,50,100]` | Custom ranges | `[0-50, 50-100]` (2 bins) |
| `'auto'` | Automatic (recommended) | Python decides optimal bins |

**Pro Tip:** More bins = More detail, but too many makes it messy. Start with 10-20 bins.

---

## üíª Professional Implementation

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data: 100 student exam scores
scores = [45, 67, 89, 52, 78, 55, 61, 73, 82, 91, 48, 76,
          88, 63, 70, 58, 85, 92, 44, 68, ...]  # 100 scores

# Create histogram
plt.hist(scores,
         bins=5,
         color='#45B7D1',
         edgecolor='black',
         alpha=0.7,
         range=(0, 100))

# Styling
plt.xlabel('Score Range')
plt.ylabel('Number of Students')
plt.title('Exam Score Distribution')
plt.xticks([0, 20, 40, 60, 80, 100])
plt.grid(axis='y', alpha=0.3)

# Display
plt.show()
```

---

## üéØ Histogram vs Bar Chart

| Feature | Histogram | Bar Chart |
|---------|-----------|-----------|
| **Data Type** | Continuous numbers | Discrete categories |
| **X-axis** | Number ranges (bins) | Category names |
| **Purpose** | Show distribution pattern | Compare categories |
| **Example** | Student ages, heights, scores | Products, cities, months |
| **Gaps** | No gaps between bars | Gaps between bars |

---

## üìä Common Bin Configurations

| Use Case | Bins | Reason |
|----------|------|--------|
| **Quick overview** | `5-10` | Broad patterns |
| **Detailed analysis** | `20-30` | Fine-grained view |
| **Age groups** | `[0,18,35,50,65,100]` | Custom age ranges |
| **Grades** | `[0,40,50,60,70,80,90,100]` | Grade boundaries |

---

## ‚ö†Ô∏è Best Practices

‚úÖ Use for **continuous numerical data**  
‚úÖ Choose appropriate `bins` (not too few, not too many)  
‚úÖ Add `edgecolor='black'` for clarity  
‚úÖ Use `alpha=0.7` when overlapping histograms  
‚úÖ Always label axes clearly  
‚ùå Don't use for categorical data (use bar chart)  
‚ùå Avoid too many bins (makes it noisy)

````python
# Sample data: 100 student exam scores
scores = [45, 67, 89, 52, 78, 55, 61, 73, 82, 91, 48, 76,
          88, 63, 70, 58, 85, 92, 44, 68]  # 100 scores

# Create histogram
plt.hist(scores,
         bins=5,
         color='#45B7D1',
         edgecolor='black',
         alpha=0.7,
         range=(0, 100))

# Styling
plt.xlabel('Score Range')
plt.ylabel('Number of Students')
plt.title('Exam Score Distribution')
plt.xticks([0, 20, 40, 60, 80, 100])
plt.grid(axis='y', alpha=0.3)

# Display
plt.show()
````

# üîµ Scatter Plot - Complete Guide

## What is a Scatter Plot?

A **scatter plot** shows the relationship between two numerical variables. Each point represents one observation with two values (x, y). It helps identify patterns, correlations, and clusters.

**Simple Rule:** Points pattern reveals relationship üìç

---

## üéØ Why Use Scatter Plots?

**Problem:** Does studying more hours lead to better exam scores?

**Solution:** Scatter plot shows if there's a relationship:
- **Positive:** More hours ‚Üí Higher scores (upward trend)
- **Negative:** More hours ‚Üí Lower scores (downward trend)
- **No pattern:** Hours don't affect scores (random scatter)

---

## üì¶ Sample Dataset

**Student Study Hours vs Exam Scores**

| Student | Study Hours (X) | Exam Score (Y) |
|---------|-----------------|----------------|
| 1 | 2 | 45 |
| 2 | 4 | 62 |
| 3 | 6 | 71 |
| 4 | 3 | 55 |
| 5 | 8 | 85 |
| 6 | 5 | 68 |
| 7 | 7 | 78 |
| 8 | 1 | 38 |

**Each point = One student with (hours, score)**

---

## üîß Syntax Breakdown

```python
plt.scatter(x, y, s, c, marker, alpha, edgecolors, label)
```

### Parameters Explained

| Parameter | Type | Required | Description | Example |
|-----------|------|----------|-------------|---------|
| `x` | array/list | ‚úÖ **Yes** | X-axis values (study hours) | `[2, 4, 6, 3, 8, 5, 7, 1]` |
| `y` | array/list | ‚úÖ **Yes** | Y-axis values (exam scores) | `[45, 62, 71, 55, 85, 68, 78, 38]` |
| `s` | int/array | ‚ùå No | Point size (default: 20) | `50` (medium), `100` (large) |
| `c` | string/array | ‚ùå No | Point color(s) | `'blue'`, `'#FF6B6B'` |
| `marker` | string | ‚ùå No | Point shape (default: 'o') | `'o'` (circle), `'^'` (triangle), `'s'` (square) |
| `alpha` | float | ‚ùå No | Transparency (0=invisible, 1=solid) | `0.6` (semi-transparent) |
| `edgecolors` | string | ‚ùå No | Point border color | `'black'`, `'white'` |
| `label` | string | ‚ùå No | Legend label | `'Student Data'` |

---

## üé® Understanding `marker`

**marker** determines the shape of each point.

### Common Marker Styles

| Marker | Shape | Use Case |
|--------|-------|----------|
| `'o'` | ‚óè Circle | Default, general use |
| `'s'` | ‚ñ† Square | Categorical grouping |
| `'^'` | ‚ñ≤ Triangle | Pointing upward trends |
| `'*'` | ‚òÖ Star | Highlighting outliers |
| `'+'` | ‚úö Plus | Intersection points |
| `'x'` | ‚úï Cross | Negative/errors |
| `'D'` | ‚óÜ Diamond | Special observations |

---

## üíª Professional Implementation

```python
import matplotlib.pyplot as plt

# Data
study_hours = [2, 4, 6, 3, 8, 5, 7, 1]
exam_scores = [45, 62, 71, 55, 85, 68, 78, 38]

# Create scatter plot
plt.scatter(study_hours,
            exam_scores,
            s=100,
            c='#45B7D1',
            marker='o',
            alpha=0.7,
            edgecolors='black',
            label='Students')

# Styling
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.title('Study Hours vs Exam Performance')
plt.grid(True, alpha=0.3)
plt.legend()

# Display
plt.show()
```

---

## üîç Identifying Patterns

### Types of Relationships

| Pattern | Visual | Interpretation |
|---------|--------|----------------|
| **Positive** | ‚ÜóÔ∏è Upward slope | Both variables increase together |
| **Negative** | ‚ÜòÔ∏è Downward slope | One increases, other decreases |
| **No correlation** | Random scatter | No relationship between variables |
| **Clustering** | Groups of points | Distinct categories/groups exist |

---

## üéØ Scatter vs Other Charts

| Feature | Scatter Plot | Line Chart | Bar Chart |
|---------|--------------|------------|-----------|
| **Shows** | Relationship | Trend over time | Category comparison |
| **Data Type** | Two numerical | Sequential | Categorical + numerical |
| **Points** | Individual observations | Connected | No points |
| **Best For** | Correlation analysis | Time series | Comparing groups |

---

## üìä Advanced: Point Sizing

You can vary point size based on a third variable:

```python
# Point size represents number of students with same values
sizes = [50, 100, 150, 75, 200, 125, 175, 50]

plt.scatter(study_hours, exam_scores, s=sizes, alpha=0.6)
```

**Result:** Bigger points = More students with those values

---

## ‚ö†Ô∏è Best Practices

‚úÖ Use for **two continuous numerical variables**  
‚úÖ Check for **correlation patterns** (upward/downward trends)  
‚úÖ Identify **outliers** (points far from the pattern)  
‚úÖ Use `alpha=0.6` when points overlap  
‚úÖ Add `edgecolors` for better visibility  
‚úÖ Look for **clusters** (grouped points)  
‚ùå Don't connect points with lines (unless time series)  
‚ùå Avoid when one variable is categorical

````python
# Data
study_hours = [2, 4, 6, 3, 8, 5, 7, 1]
exam_scores = [45, 62, 71, 55, 85, 68, 78, 38]

# Create scatter plot
plt.scatter(study_hours,
            exam_scores,
            s=100,
            c='#45B7D1',
            marker='^',
            alpha=0.7,
            edgecolors='black',
            label='Students')

# Styling
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.title('Study Hours vs Exam Performance')
plt.grid(True, alpha=0.3)
plt.legend()

# Display
plt.show()
````

# üé® Subplots - Complete Guide

## What are Subplots?

**Subplots** allow you to display multiple charts on the same page/canvas. Instead of creating separate figures, you divide one canvas into sections (grid layout).

**Simple Rule:** Multiple charts, one canvas üìä‚ûïüìà

---

## üéØ Why Use Subplots?

### Problem
You have:
- Line chart for **Sales** over time
- Bar chart for **Expenses** by category

**Challenge:** Need to compare both, but they're on separate pages!

### Solution: Subplots

**Benefits:**
- ‚úÖ **Compare side-by-side** - Easy visual comparison
- ‚úÖ **Quick analysis** - View related datasets together
- ‚úÖ **Save space** - One canvas instead of multiple
- ‚úÖ **Professional reports** - Clean, organized layout

**Example Grid:**
```

‚îÇ   Chart 1   ‚îÇ   Chart 2   ‚îÇ
‚îÇ   (Sales)   ‚îÇ (Expenses)  ‚îÇ
‚îî‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚î¥‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îò
    1 Canvas, 2 Sections
```

---

## üîß Method 1: Quick Subplot Creation

### Syntax

```python
plt.subplot(nrows, ncols, index)
```

### Parameters Explained

| Parameter | Type | Required | Description | Example |
|-----------|------|----------|-------------|---------|
| `nrows` | int | ‚úÖ **Yes** | Number of rows in grid | `2` (2 rows) |
| `ncols` | int | ‚úÖ **Yes** | Number of columns in grid | `3` (3 columns) |

# üíæ Saving Figures - Complete Guide

## What is `savefig()`?

**`savefig()`** saves your matplotlib charts as image files instead of just displaying them. This allows you to use charts in reports, presentations, websites, or share them with others.

**Simple Rule:** Create once, save forever üì∏

---

## üéØ Why Save Figures?

### Use Cases

| Scenario | Benefit |
|----------|---------|
| **Reports & Documents** | Insert charts into Word/PDF documents |
| **Presentations** | Add visualizations to PowerPoint slides |
| **Websites & Blogs** | Display charts on web pages |
| **Email & Sharing** | Send analysis results to colleagues |
| **Archiving** | Keep records of analysis over time |
| **Publications** | Submit high-quality charts for papers |

**Problem:** `plt.show()` only displays, doesn't save  
**Solution:** Use `plt.savefig()` to export as files

---

## üîß Syntax Breakdown

```python
plt.savefig(filename, dpi, format, bbox_inches, transparent, facecolor, pad_inches)
```

### Parameters Explained

|-----------|------|----------|-------------|---------|
| `filename` | string | ‚úÖ **Yes** | File path with extension | `'sales_chart.png'` |

## üé® What is `ax`?

**`ax`** (short for "axes") is a **container** that holds your subplot charts.

**Think of it like:**
```
fig = The entire canvas (frame)
ax  = Individual painting spots on the canvas
```

### Single Row/Column: `ax` is a 1D list

```python
fig, ax = plt.subplots(1, 3)  # 1 row, 3 columns
```

Access like a list:
- `ax[0]` = first subplot
- `ax[1]` = second subplot  
- `ax[2]` = third subplot

### Multiple Rows & Columns: `ax` is a 2D array

```python
fig, ax = plt.subplots(2, 2)  # 2 rows, 2 columns
```

Access like a table:
- `ax[0, 0]` = row 0, column 0
- `ax[0, 1]` = row 0, column 1
- `ax[1, 0]` = row 1, column 0
- `ax[1, 1]` = row 1, column 1

## üéØ Key Methods for Each Subplot

Instead of using `plt.something()`, use `ax[position].something()`:

| Regular Plot | With Subplots |
|--------------|---------------|
| `plt.plot(x, y)` | `ax[0].plot(x, y)` |
| `plt.bar(x, y)` | `ax[1].bar(x, y)` |
| `plt.scatter(x, y)` | `ax[0, 0].scatter(x, y)` |
| `plt.hist(data)` | `ax[0, 1].hist(data)` |
| `plt.title('Title')` | `ax[0].set_title('Title')` |
| `plt.xlabel('X')` | `ax[0].set_xlabel('X')` |
| `plt.ylabel('Y')` | `ax[0].set_ylabel('Y')` |
| `plt.legend()` | `ax[0].legend()` |
| `plt.grid(True)` | `ax[0].grid(True)` |

**Notice:** `plt.something()` becomes `ax[position].something()` (add `set_` for labels/title)

## ‚ö†Ô∏è Common Mistakes & Solutions

### Mistake 1: Using `plt` instead of `ax`

```python
# ‚ùå Wrong
fig, ax = plt.subplots(1, 2)
plt.plot(x, y)  # Will plot in last active subplot only!

# ‚úÖ Correct
fig, ax = plt.subplots(1, 2)
ax[0].plot(x, y)  # Plots in specific subplot
```

---

### Mistake 2: Wrong index for 2D grid

```python
fig, ax = plt.subplots(2, 2)

# ‚ùå Wrong
ax[0].plot(x, y)  # Error! Need 2 indices

# ‚úÖ Correct
ax[0, 0].plot(x, y)  # row, column
```

---

### Mistake 3: Forgetting `tight_layout()`

```python
# ‚ùå Without tight_layout
fig, ax = plt.subplots(1, 2)
ax[0].plot(x, y)
ax[1].bar(x, y)
plt.show()  # Overlapping titles/labels!

# ‚úÖ With tight_layout
fig, ax = plt.subplots(1, 2)
ax[0].plot(x, y)
ax[1].bar(x, y)
plt.tight_layout()  # Clean spacing!
plt.show()
```
